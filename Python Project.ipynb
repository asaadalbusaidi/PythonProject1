```python
#Q1 The 1st quartile/25th percentile of the data.
#Q2 The 2nd quartile/median/50th percentile of the data.
#Q3 The 3rd quartile/75th percentile of the data.
out=[]
def iqr_outliers(df):
    q1 = df.quantile(0.25)
    q3 = df.quantile(0.75)
    iqr = q3-q1
    Lower_tail = q1 - 1.5 * iqr
    Upper_tail = q3 + 1.5 * iqr
    for i in df:
        if i > Upper_tail or i < Lower_tail:
            out.append(i)
    print("Outliers:",out)
iqr_outliers(df['NumCompaniesWorked'])
iqr_outliers(df['MonthlyIncome'])
iqr_outliers(df['StockOptionLevel'])
iqr_outliers(df['TotalWorkingYears'])
iqr_outliers(df['TrainingTimesLastYear'])
iqr_outliers(df['YearsWithCurrManager'])
```

```python
# Droping the ouliers using percentile
def drop_outliers(df, field_name):
    iqr = 1.5 * (np.percentile(df[field_name], 75) - np.percentile(df[field_name], 25))
    df.drop(df[df[field_name] > (iqr + np.percentile(df[field_name], 75))].index, inplace=True)
    df.drop(df[df[field_name] < (np.percentile(df[field_name], 25) - iqr)].index, inplace=True)
drop_outliers(df, 'MonthlyIncome')
drop_outliers(df, 'StockOptionLevel')
drop_outliers(df, 'TotalWorkingYears')
drop_outliers(df, 'TrainingTimesLastYear')
drop_outliers(df, 'YearsWithCurrManager')
drop_outliers(df, 'NumCompaniesWorked')
```
